```python
import geopandas as gpd
from libpysal.weights import Queen
from esda import Moran, Moran_Local
from splot.esda import plot_moran, lisa_cluster
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches


significance_level = 0.05
cluster_colors = {
    'HH': '#D6191C',   # High-High - Red
    'HL': '#FCAD61',   # High-Low - Orange
    'LH': '#AAD9E9',   # Low-High - Light Blue
    'LL': '#2C7AB6',   # Low-Low - Blue
    'ns': '#D3D3D3'    # Not significant - Gray
}
```

```python
variables = {'Education': 'Higher_Education_Proportion', 'Income': 'median_income', 'Home Value': 'median_value'}
years = merged_gdf['year'].unique()
```

```python
def calculate_morans_i(year_data, var_name, var_col):
    weights = Queen.from_dataframe(year_data)
    weights.transform = 'R'
    y = year_data[var_col]
    moran = Moran(y, weights)
    print(f"{var_name} Moran's I for Missouri in {year_data['year'].iloc[0]}: {moran.I}, p-value: {moran.p_sim}")
    
    plt.figure(figsize=(10, 5))
    plot_moran(moran, zstandard=False, fitline_kwds=dict(color='#4393c3'))
    plt.title(f"Moran's I Plot for {var_name} ({year_data['year'].iloc[0]}) - Missouri")
    plt.show()
    
    return weights, moran

def calculate_local_morans_i(year_data, var_name, var_col, weights):
    y = year_data[var_col]
    local_moran = Moran_Local(y, weights)
    year_data[f'sig_{var_name.lower()}_cluster'] = np.where(
        local_moran.p_sim < significance_level, local_moran.q, 5  # Not significant
    )
    cluster_labels = {1: 'HH', 2: 'LH', 3: 'LL', 4: 'HL', 5: 'ns'}
    year_data[f'sig_{var_name.lower()}_cluster'] = year_data[f'sig_{var_name.lower()}_cluster'].map(cluster_labels)
    print(f"{var_name} Cluster Counts in Missouri for {year_data['year'].iloc[0]}:\n", year_data[f'sig_{var_name.lower()}_cluster'].value_counts())
    return local_moran

def plot_missouri_clusters(local_moran, year_data, var_name):
    fig, ax = plt.subplots(1, 1, figsize=(15, 10))
    lisa_cluster(local_moran, year_data, p=significance_level, ax=ax)
    ax.set_title(f"{var_name} Local Moran's I Clusters - Missouri ({year_data['year'].iloc[0]})")
    plt.show()

def plot_greene_county(year_data, var_name):
    greene_county_data = year_data[year_data['COUNTYFP'] == '077'].copy()
    print(f"{var_name} Cluster Counts in Greene County for {year_data['year'].iloc[0]}:\n", greene_county_data[f'sig_{var_name.lower()}_cluster'].value_counts())
    greene_county_data['color'] = greene_county_data[f'sig_{var_name.lower()}_cluster'].map(cluster_colors)
    
    fig, ax = plt.subplots(1, 1, figsize=(15, 10))
    greene_county_data.plot(color=greene_county_data['color'], edgecolor='#E2E2E2', linewidth=0.8, ax=ax)
    ax.axis('off')
    
    legend_patches = [
        mpatches.Patch(color=color, label=label)
        for label, color in cluster_colors.items()
    ]
    # ax.legend(handles=legend_patches, loc='upper right', title=f"{var_name} Cluster ({year_data['year'].iloc[0]}) - Greene County", frameon=False)
    plt.show()
```

## Moran's I Table Creation

```python
import pandas as pd
import geopandas as gpd
from libpysal.weights import Queen
from esda import Moran, Moran_Local
import numpy as np
from splot.esda import lisa_cluster
import matplotlib.pyplot as plt


# Create a new column for the original (non-stripped) version
df_merged_2['geo_id_original'] = df_merged_2['geo_id']

df_merged_2['geo_id_stripped'] = df_merged_2['geo_id'].str[-12:]

# Rename the column for mergin purposes
df_merged_2['GEOID'] = df_merged_2['geo_id_stripped']

merged_gdf = block_groups.merge(df_merged_2, on='GEOID')
merged_gdf['geo_id'] = df_merged_2['geo_id']


# Reproject for spatial weights calculation
merged_gdf = merged_gdf.to_crs(epsg=3857)

education_var = 'Higher_Education_Proportion'
income_var = 'median_income'
home_var = 'median_value'
significance_level = 0.05

yearly_dfs = []

# Create Moran's I for each variable
for year in merged_gdf['year'].unique():
    yearly_data = merged_gdf[merged_gdf['year'] == year].copy()

    weights = Queen.from_dataframe(yearly_data)
    weights.transform = 'R'
    
    for var, label in [(education_var, 'education'), (income_var, 'income'), (home_var, 'home_value')]:
        y = yearly_data[var]
        moran_local = Moran_Local(y, weights)
        
        yearly_data[f'{label}_cluster'] = np.where(
            moran_local.p_sim < significance_level,
            moran_local.q,
            5  # Not significant
        )
        
        yearly_data[f'{label}_cluster'] = yearly_data[f'{label}_cluster'].map({1: 'HH', 2: 'LH', 3: 'LL', 4: 'HL', 5: 'ns'})
    
    yearly_data = yearly_data[['geo_id', 'year', 'education_cluster', 'income_cluster', 'home_value_cluster', 'geometry']]
    yearly_dfs.append(yearly_data)

final_gdf = gpd.GeoDataFrame(pd.concat(yearly_dfs), geometry='geometry')

final_gdf.columns = ['geo_id', 'year', 'education_moran_quad', 'income_moran_quad', 'home_value_moran_quad', 'geometry']
final_gdf.reset_index(drop=True, inplace=True)

final_gdf.info()
```

```python
#DB CONNECTION

import psycopg2
from psycopg2 import sql
from psycopg2.extensions import adapt, register_adapter, AsIs
from sqlalchemy import create_engine,text
import getpass

import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

mypasswd = getpass.getpass("Enter your database password: ")
```

```python
final_gdf = final_gdf.drop(columns=['geometry']) # necessary to drop due to lack of PostGIS extension
```

```python
final_gdf = final_gdf.dropna(subset=['geo_id'])
```

```python
final_gdf.shape
```

```python
final_gdf = final_gdf.drop_duplicates(subset=['geo_id', 'year'])
```

```python
final_gdf.shape
```

```python
try:
    conn = psycopg2.connect(database=database, user=user, host=host, password=password)
    cursor = conn.cursor()
    print("Connected to the database.")
    
    cursor.execute("DROP TABLE IF EXISTS moran_clusters;")
    print("Dropped the moran_clusters table if it existed.")
    
    create_table_query = """
    CREATE TABLE moran_clusters (
        geo_id TEXT,
        year INTEGER,
        education_moran_quad TEXT,
        income_moran_quad TEXT,
        home_value_moran_quad TEXT,
        PRIMARY KEY (geo_id, year)
    );
    """
    cursor.execute(create_table_query)
    conn.commit()
    print("Created the moran_clusters table.")
    
    chunk_size = 500
    for start in range(0, len(final_gdf), chunk_size):
        chunk = final_gdf.iloc[start:start + chunk_size]
        chunk.columns = [col.lower() for col in chunk.columns]
        chunk.to_sql('moran_clusters', engine, if_exists='append', index=False)
        print(f"Uploaded a chunk of {len(chunk)} records to the moran_clusters table.")
    
    print("Data uploaded successfully.")
    
    query = "SELECT * FROM moran_clusters;"
    moran_df = pd.read_sql(query, conn)
    print("Data extracted successfully! DataFrame shape:", moran_df.shape)

except Exception as e:
    print("An error occurred:", e)

finally:
    if cursor:
        cursor.close()
    if conn:
        conn.close()
```
