```python
import psycopg2
from psycopg2 import sql
import numpy as np
import pandas as pd
from psycopg2.extensions import adapt, register_adapter, AsIs
from sqlalchemy import create_engine, text
import getpass
```

```python
file_paths = ['2018_Employment_Status.csv', 
              '2019_Employment_Status.csv',
              '2020_Employment_Status.csv', 
              '2021_Employment_Status.csv', 
              '2022_Employment_Status.csv']

dataframes = [pd.read_csv(file) for file in file_paths]
Employment_Status = pd.concat(dataframes, ignore_index=True)
merged_file_path = 'Employment_Status.csv'
Employment_Status.to_csv(merged_file_path, index=False)
merged_file_path
```

```python
database = "f24t03"
user     = input("Type username (pawprint) and hit enter: ")
password = getpass.getpass("Type password and hit enter: ")

connection = psycopg2.connect(database = database,
                              user     = user,
                              host     = 'pgsql',
                              password = password)
```

```python
with connection, connection.cursor() as cursor:
    cursor.execute("DROP TABLE IF EXISTS employment_status;")
    

    cursor.execute(
        '''
        CREATE TABLE employment_status (
    geo_id TEXT,
    block_group TEXT,
    census_tract TEXT,
    county TEXT,
    state TEXT,
    est_total INTEGER,
    moe_total INTEGER,
    est_tot_in_labor_force INTEGER,
    moe_tot_in_labor_force INTEGER,
    est_tot_civilian INTEGER,
    moe_tot_civilian INTEGER,
    est_tot_civilian_labor_employed INTEGER,
    moe_tot_civilian_labor_employed INTEGER,
    est_tot_civilian_unemployed INTEGER,
    moe_tot_civilian_unemployed INTEGER,
    est_tot_armed_forces INTEGER,
    moe_tot_armed_forces INTEGER,
    est_total_not_in_labor_force INTEGER,
    moe_total_not_in_labor_force INTEGER,
    year INTEGER,
    PRIMARY KEY (geo_id, year)
    );
    '''
    )
```
