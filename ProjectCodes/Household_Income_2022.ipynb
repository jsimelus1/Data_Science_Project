```python
import pandas as pd

data = pd.read_csv("ACSDT5Y2022.B19001-Data.csv")
data.head()
```

```python
data[['BLOCK_GROUP', 'CENSUS_TRACT', 'COUNTIES', 'STATE']] = data['NAME'].str.split(';', expand=True)
data.drop(columns=['NAME'], inplace=True)
data['BLOCK_GROUP'] = data['BLOCK_GROUP'].str.strip()
data['CENSUS_TRACT'] = data['CENSUS_TRACT'].str.replace('County', '').str.strip()
data['COUNTIES'] = data['COUNTIES'].str.strip()
data['STATE'] = data['STATE'].str.strip()
data.head()
```

```python
data = data[['GEO_ID', 'BLOCK_GROUP', 'CENSUS_TRACT', 'COUNTIES', 'STATE'] + 
            [col for col in data.columns if col not in ['GEO_ID', 'BLOCK_GROUP', 'CENSUS_TRACT',
                                                        'COUNTIES', 'STATE']]]
data.head()
```

```python
new_headers = data.iloc[0, 5:].tolist()
new_headers = ['GEO_ID', 'BLOCK_GROUP', 'CENSUS_TRACT', 'COUNTIES', 'STATE'] + new_headers
data.columns = new_headers

data = data.drop(index=0)
data.head()
```

```python
data.columns = data.columns.astype(str)
def clean_column_name(column_name):
    column_name = column_name.replace("!!", "_")
    column_name = column_name.replace(" ", "_")
    column_name = column_name.lower()
    column_name = ''.join(char if char.isalnum() or char == '_' else '' for char in column_name)
    return column_name

cleaned_column_names = [clean_column_name(col) for col in data.columns]
data.columns = cleaned_column_names
data.drop(columns=['nan'], inplace=True)
data.head()
```

```python
column_mapping = {
    'geo_id': 'geo_id',
    'block_group': 'block_group',
    'census_tract': 'census_tract',
    'counties': 'county',
    'state': 'state',
    'estimate_total': 'est_total',
    'margin_of_error_total': 'moe_total',
    'estimate_total_less_than_10000': 'est_less_than_10k',
    'margin_of_error_total_less_than_10000': 'moe_less_than_10k',
    'estimate_total_10000_to_14999': 'est_10to15k',
    'margin_of_error_total_10000_to_14999': 'moe_10to15k',
    'estimate_total_15000_to_19999': 'est_15to20k',
    'margin_of_error_total_15000_to_19999': 'moe_15to20k',
    'estimate_total_20000_to_24999': 'est_20to25k',
    'margin_of_error_total_20000_to_24999': 'moe_20to25k',
    'estimate_total_25000_to_29999': 'est_25to30k',
    'margin_of_error_total_25000_to_29999': 'moe_25to30k',
    'estimate_total_30000_to_34999': 'est_30to35k',
    'margin_of_error_total_30000_to_34999': 'moe_30to35k',
    'estimate_total_35000_to_39999': 'est_35to40k',
    'margin_of_error_total_35000_to_39999': 'moe_35to40k',
    'estimate_total_40000_to_44999': 'est_40to45k',
    'margin_of_error_total_40000_to_44999': 'moe_40to45k',
    'estimate_total_45000_to_49999': 'est_45to50k',
    'margin_of_error_total_45000_to_49999': 'moe_45to50k',
    'estimate_total_50000_to_59999': 'est_50to60k',
    'margin_of_error_total_50000_to_59999': 'moe_50to60k',
    'estimate_total_60000_to_74999': 'est_60to75k',
    'margin_of_error_total_60000_to_74999': 'moe_60to75k',
    'estimate_total_75000_to_99999': 'est_75to100k',
    'margin_of_error_total_75000_to_99999': 'moe_75to100k',
    'estimate_total_100000_to_124999': 'est_100to125k',
    'margin_of_error_total_100000_to_124999': 'moe_100to125k',
    'estimate_total_125000_to_149999': 'est_125to150k',
    'margin_of_error_total_125000_to_149999': 'moe_125to150k',
    'estimate_total_150000_to_199999': 'est_150to200k',
    'margin_of_error_total_150000_to_199999': 'moe_150to200k',
    'estimate_total_200000_or_more': 'est_more_than_200k',
    'margin_of_error_total_200000_or_more': 'moe_more_than_200k'
}

data.rename(columns=column_mapping, inplace=True)
data.replace(['-', '**', '10,000-', '***'], '', inplace=True)
data.head()
```
