```python
import pandas as pd
import numpy as np

data = pd.read_csv("ACSDT5Y2020.B25036-Data.csv")
data.head()
```

```python
# Splitting the 'NAME' column by commas
data[['BLOCK_GROUP', 'CENSUS_TRACT', 'COUNTIES', 'STATE']] = data['NAME'].str.split(',', expand=True)

# Dropping the 'NAME' column
data.drop(columns=['NAME'], inplace=True)

# Cleaning up the GEOGRAPHIC_AREA_NAME, COUNTIES, and STATE columns
data['BLOCK_GROUP'] = data['BLOCK_GROUP'].str.strip()
data['CENSUS_TRACT'] = data['CENSUS_TRACT'].str.replace('County', '').str.strip()
data['COUNTIES'] = data['COUNTIES'].str.strip()
data['STATE'] = data['STATE'].str.strip()

# Display the first few rows
data.head()
```

```python
# Reorder the new columns 'TRACT', 'COUNTY', 'STATE' to be the second, third, and fourth columns
data = data[['GEO_ID', 'BLOCK_GROUP', 'CENSUS_TRACT', 'COUNTIES', 'STATE'] + 
            [col for col in data.columns if col not in ['GEO_ID', 'BLOCK_GROUP', 'CENSUS_TRACT',
                                                        'COUNTIES', 'STATE']]]
# Display the first few rows
data.head()
```

```python
# Save the original headers
new_headers = data.iloc[0, 5:].tolist()
new_headers = ['GEO_ID', 'BLOCK_GROUP', 'CENSUS_TRACT', 'COUNTIES', 'STATE'] + new_headers

# Replace the headers of the dataframe
data.columns = new_headers

# Drop the first row, as it's now part of the headers
data = data.drop(index=0)

# Display the first few rows of the updated dataset
data.head()
```

```python
# Define identifier columns
identifier_columns = ['GEO_ID', 'BLOCK_GROUP', 'CENSUS_TRACT', 'COUNTIES', 'STATE']

# Extract identifier columns from the original data
if all(col in data.columns for col in identifier_columns):
    identifier_data = data[identifier_columns]
else:
    raise ValueError("One or more identifier columns are missing in the original data.")

# Ensure numeric conversion for applicable columns
data.iloc[:, len(identifier_columns):] = data.iloc[:, len(identifier_columns):].apply(pd.to_numeric, errors='coerce')

# Combine columns for 'Estimate' and 'Margin of Error' for 2014 or later and 2010 to 2013
combine_columns = [
    'Estimate!!Total:!!Owner occupied:!!Built 2014 or later',
    'Estimate!!Total:!!Owner occupied:!!Built 2010 to 2013',
    'Margin of Error!!Total:!!Owner occupied:!!Built 2014 or later',
    'Margin of Error!!Total:!!Owner occupied:!!Built 2010 to 2013',
    'Estimate!!Total:!!Renter occupied:!!Built 2014 or later',
    'Margin of Error!!Total:!!Renter occupied:!!Built 2014 or later',
    'Estimate!!Total:!!Renter occupied:!!Built 2010 to 2013',
    'Margin of Error!!Total:!!Renter occupied:!!Built 2010 to 2013'
]

if all(col in data.columns for col in combine_columns[:2]):
    data['Estimate!!Total:!!Built 2010 to 2019'] = (
        data['Estimate!!Total:!!Owner occupied:!!Built 2014 or later'] 
        + data['Estimate!!Total:!!Owner occupied:!!Built 2010 to 2013']
    )
if all(col in data.columns for col in combine_columns[2:]):
    data['Margin of Error!!Total:!!Built 2010 to 2019'] = (
        data['Margin of Error!!Total:!!Owner occupied:!!Built 2014 or later'] 
        + data['Margin of Error!!Total:!!Owner occupied:!!Built 2010 to 2013']
    )
if all(col in data.columns for col in combine_columns[:2]):
    data['Estimate!!Total:!!Renter occupied:!!Built 2010 to 2019'] = (
        data['Estimate!!Total:!!Renter occupied:!!Built 2014 or later'] 
        + data['Estimate!!Total:!!Renter occupied:!!Built 2010 to 2013']
    )
if all(col in data.columns for col in combine_columns[2:]):
    data['Margin of Error!!Total:!!Renter occupied:!!Built 2010 to 2019'] = (
        data['Margin of Error!!Total:!!Renter occupied:!!Built 2014 or later'] 
        + data['Margin of Error!!Total:!!Renter occupied:!!Built 2010 to 2013']
    )

# Drop columns used in the combination
data.drop(columns=combine_columns, inplace=True, errors='ignore')

# Add new columns for 'Estimate!!Total:!!Built 2020 or later' and 'Margin of Error!!Total:!!Built 2020 or later'
data['Estimate!!Total:!!Built 2020 or later'] = 0
data['Margin of Error!!Total:!!Built 2020 or later'] = 0
data['Estimate!!Total:!!Renter occupied:!!Built 2020 or later'] = 0
data['Margin of Error!!Total:!!Renter occupied:!!Built 2020 or later'] = 0

# Merge identifier columns back into the processed DataFrame
data.reset_index(drop=True, inplace=True)
identifier_data.reset_index(drop=True, inplace=True)
data = pd.concat([identifier_data, data], axis=1)

# Reorder columns to match the specified order
ordered_columns = identifier_columns + [
    'Estimate!!Total:', 'Margin of Error!!Total:','Estimate!!Total:!!Owner occupied:',
    'Margin of Error!!Total:!!Owner occupied:', 'Estimate!!Total:!!Built 2020 or later',
    'Margin of Error!!Total:!!Built 2020 or later',
    'Estimate!!Total:!!Built 2010 to 2019', 'Margin of Error!!Total:!!Built 2010 to 2019',
    'Estimate!!Total:!!Owner occupied:!!Built 2000 to 2009',
    'Margin of Error!!Total:!!Owner occupied:!!Built 2000 to 2009',
    'Estimate!!Total:!!Owner occupied:!!Built 1990 to 1999',
    'Margin of Error!!Total:!!Owner occupied:!!Built 1990 to 1999',
    'Estimate!!Total:!!Owner occupied:!!Built 1980 to 1989',
    'Margin of Error!!Total:!!Owner occupied:!!Built 1980 to 1989',
    'Estimate!!Total:!!Owner occupied:!!Built 1970 to 1979',
    'Margin of Error!!Total:!!Owner occupied:!!Built 1970 to 1979',
    'Estimate!!Total:!!Owner occupied:!!Built 1960 to 1969',
    'Margin of Error!!Total:!!Owner occupied:!!Built 1960 to 1969',
    'Estimate!!Total:!!Owner occupied:!!Built 1950 to 1959',
    'Margin of Error!!Total:!!Owner occupied:!!Built 1950 to 1959',
    'Estimate!!Total:!!Owner occupied:!!Built 1940 to 1949',
    'Margin of Error!!Total:!!Owner occupied:!!Built 1940 to 1949',
    'Estimate!!Total:!!Owner occupied:!!Built 1939 or earlier',
    'Margin of Error!!Total:!!Owner occupied:!!Built 1939 or earlier',
    'Estimate!!Total:!!Renter occupied:',
    'Margin of Error!!Total:!!Renter occupied:',
    'Estimate!!Total:!!Renter occupied:!!Built 2020 or later',
    'Margin of Error!!Total:!!Renter occupied:!!Built 2020 or later',
    'Estimate!!Total:!!Renter occupied:!!Built 2010 to 2019',
    'Margin of Error!!Total:!!Renter occupied:!!Built 2010 to 2019',
    'Estimate!!Total:!!Renter occupied:!!Built 2000 to 2009',
    'Margin of Error!!Total:!!Renter occupied:!!Built 2000 to 2009',
    'Estimate!!Total:!!Renter occupied:!!Built 1990 to 1999',
    'Margin of Error!!Total:!!Renter occupied:!!Built 1990 to 1999',
    'Estimate!!Total:!!Renter occupied:!!Built 1980 to 1989',
    'Margin of Error!!Total:!!Renter occupied:!!Built 1980 to 1989',
    'Estimate!!Total:!!Renter occupied:!!Built 1970 to 1979',
    'Margin of Error!!Total:!!Renter occupied:!!Built 1970 to 1979',
    'Estimate!!Total:!!Renter occupied:!!Built 1960 to 1969',
    'Margin of Error!!Total:!!Renter occupied:!!Built 1960 to 1969',
    'Estimate!!Total:!!Renter occupied:!!Built 1950 to 1959',
    'Margin of Error!!Total:!!Renter occupied:!!Built 1950 to 1959',
    'Estimate!!Total:!!Renter occupied:!!Built 1940 to 1949',
    'Margin of Error!!Total:!!Renter occupied:!!Built 1940 to 1949',
    'Estimate!!Total:!!Renter occupied:!!Built 1939 or earlier',
    'Margin of Error!!Total:!!Renter occupied:!!Built 1939 or earlier'
]

# Retain only columns that exist in the DataFrame
existing_columns = [col for col in ordered_columns if col in data.columns]

# Reorder the DataFrame columns
data = data[existing_columns]

# Display the updated DataFrame to verify
print("Updated DataFrame with identifier columns:")
data.head()
```

```python
# Identify and drop duplicate columns
data = data.loc[:, ~data.columns.duplicated()]

# Display the updated DataFrame
print("DataFrame after removing duplicate columns:")
print(data.head())

# Print the updated column names to verify
print("Updated column names:")
print(data.columns)
```

```python
# Let's clean and standardize the column names
def clean_column_name(column_name):
    # Convert the column name to a string first
    column_name = str(column_name)
    # Replace "!!" with "_"
    column_name = column_name.replace("!!", "_")
    # Replace spaces with underscores
    column_name = column_name.replace(" ", "_")
    # Convert to lowercase
    column_name = column_name.lower()
    # Remove any characters that are not alphanumeric or underscores
    column_name = ''.join(char if char.isalnum() or char == '_' else '' for char in column_name)
    return column_name

# Apply this function to all column names
cleaned_column_names = [clean_column_name(col) for col in data.columns]

# Rename the dataset columns
data.columns = cleaned_column_names

# Display the first few rows of the dataset with the new column names
data.head()
```
