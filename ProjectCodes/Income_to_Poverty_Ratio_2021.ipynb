```python
import pandas as pd
import numpy as np

data = pd.read_csv("ACSDT5Y2021.C17002-Data.csv")
data.head()
```

```python
data[['BLOCK_GROUP', 'CENSUS_TRACT', 'COUNTIES', 'STATE']] = data['NAME'].str.split(',', expand=True)
data.drop(columns=['NAME'], inplace=True)
data['BLOCK_GROUP'] = data['BLOCK_GROUP'].str.strip()
data['CENSUS_TRACT'] = data['CENSUS_TRACT'].str.replace('County', '').str.strip()
data['COUNTIES'] = data['COUNTIES'].str.strip()
data['STATE'] = data['STATE'].str.strip()
data.head()
```

```python
data = data[['GEO_ID', 'BLOCK_GROUP', 'CENSUS_TRACT', 'COUNTIES', 'STATE'] + 
            [col for col in data.columns if col not in ['GEO_ID', 'BLOCK_GROUP', 'CENSUS_TRACT',
                                                        'COUNTIES', 'STATE']]]
data.head()
```

```python
new_headers = data.iloc[0, 5:].tolist()
new_headers = ['GEO_ID', 'BLOCK_GROUP', 'CENSUS_TRACT', 'COUNTIES', 'STATE'] + new_headers
data.columns = new_headers
data = data.drop(index=0)
data.head()
```

```python
def clean_column_name(column_name):
    column_name = str(column_name)
    column_name = column_name.replace("!!", "_")
    column_name = column_name.replace(" ", "_")
    column_name = column_name.lower()
    column_name = ''.join(char if char.isalnum() or char == '_' else '' for char in column_name)
    return column_name

cleaned_column_names = [clean_column_name(col) for col in data.columns]
data.columns = cleaned_column_names
data.head()
```

```python
if 'nan' in data.columns:
    data = data.drop(columns=['nan'])
data.head()
```

```python
data.columns.tolist()
```

```python
column_mapping = {
    'geo_id': 'geo_id',
    'block_group': 'block_group',
    'census_tract': 'census_tract',
    'counties': 'county',
    'state': 'state',
    'estimate_total': 'est_total',
    'margin_of_error_total': 'moe_total',
    'estimate_total_under_50': 'est_under_50',
    'margin_of_error_total_under_50': 'moe_under_50',
    'estimate_total_50_to_99': 'est_50_to_99',
    'margin_of_error_total_50_to_99': 'moe_50_to_99',
    'estimate_total_100_to_124': 'est_100_to_124',
    'margin_of_error_total_100_to_124': 'moe_100_to_124',
    'estimate_total_125_to_149': 'est_125_to_149',
    'margin_of_error_total_125_to_149': 'moe_125_to_149',
    'estimate_total_150_to_184': 'est_150_to_184',
    'margin_of_error_total_150_to_184': 'moe_150_to_184',
    'estimate_total_185_to_199': 'est_185_to_199',
    'margin_of_error_total_185_to_199': 'moe_185_to_199',
    'estimate_total_200_and_over': 'est_200_plus',
    'margin_of_error_total_200_and_over': 'moe_200_plus'
}

data.rename(columns=column_mapping, inplace=True)
print("Column names after renaming:")
print(data.columns)

numeric_columns = [
    'est_total',
    'moe_total',
    'est_under_50',
    'moe_under_50',
    'est_50_to_99',
    'moe_50_to_99',
    'est_100_to_124',
    'moe_100_to_124',
    'est_125_to_149',
    'moe_125_to_149',
    'est_150_to_184',
    'moe_150_to_184',
    'est_185_to_199',
    'moe_185_to_199',
    'est_200_plus',
    'moe_200_plus'
]

missing_columns = [col for col in numeric_columns if col not in data.columns]
if missing_columns:
    print(f"These columns are missing after renaming: {missing_columns}")
else:
    data.replace(['**', '***'], np.nan, inplace=True)
    data[numeric_columns] = data[numeric_columns].replace({',': '', '-': np.nan, r'\+': ''}, regex=True)
    data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric, errors='raise')

data.fillna('', inplace=True)
if 'nan' in data.columns:
    data.drop(columns=['nan'], inplace=True)
data.head()
```

```python
data['year'] = 2021
data.head()
```
