```python
import psycopg2
from psycopg2 import sql
import numpy as np
import pandas as pd
from psycopg2.extensions import adapt, register_adapter, AsIs
from sqlalchemy import create_engine, text
import getpass
```

```python
file_paths = ['2018_Median_Home_Value.csv', 
              '2019_Median_Home_Value.csv', 
              '2020_Median_Home_Value.csv', 
              '2021_Median_Home_Value.csv', 
              '2022_Median_Home_Value.csv']
dataframes = [pd.read_csv(file) for file in file_paths]
Median_Home = pd.concat(dataframes, ignore_index=True)
merged_file_path = 'Median_Home_Value.csv'
Median_Home.to_csv(merged_file_path, index=False)
merged_file_path
```

```python
with connection, connection.cursor() as cursor:
    cursor.execute("DROP TABLE IF EXISTS median_home_value;")
    

    cursor.execute(
        '''
        CREATE TABLE median_home_value (
    geo_id TEXT,
    block_group TEXT,
    census_tract TEXT,
    county TEXT,
    state TEXT,
    est_median_value INTEGER,
    total_margin_of_error INTEGER,
    year INTEGER,
    PRIMARY KEY (geo_id, year)
    );
        '''
    )
```
