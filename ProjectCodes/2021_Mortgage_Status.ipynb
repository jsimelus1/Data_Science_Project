```python
import pandas as pd
import numpy as np

data = pd.read_csv("ACSDT5Y2021.B25081-Data.csv")
data.head()
```

```python
data[['BLOCK_GROUP', 'CENSUS_TRACT', 'COUNTIES', 'STATE']] = data['NAME'].str.split(',', expand=True)
data.drop(columns=['NAME'], inplace=True)
data['BLOCK_GROUP'] = data['BLOCK_GROUP'].str.strip()
data['CENSUS_TRACT'] = data['CENSUS_TRACT'].str.replace('County', '').str.strip()
data['COUNTIES'] = data['COUNTIES'].str.strip()
data['STATE'] = data['STATE'].str.strip()
data.head()
```

```python
data = data[['GEO_ID', 'BLOCK_GROUP', 'CENSUS_TRACT', 'COUNTIES', 'STATE'] + 
            [col for col in data.columns if col not in ['GEO_ID', 'BLOCK_GROUP', 'CENSUS_TRACT',
                                                        'COUNTIES', 'STATE']]]

data.head()
```

```python
new_headers = data.iloc[0, 5:].tolist()
new_headers = ['GEO_ID', 'BLOCK_GROUP', 'CENSUS_TRACT', 'COUNTIES', 'STATE'] + new_headers
data.columns = new_headers
data = data.drop(index=0)
data.head()
```

```python
def clean_column_name(column_name):
    column_name = str(column_name)
    column_name = column_name.replace("!!", "_")
    column_name = column_name.replace(" ", "_")
    column_name = column_name.lower()
    column_name = ''.join(char if char.isalnum() or char == '_' else '' for char in column_name)
    return column_name

cleaned_column_names = [clean_column_name(col) for col in data.columns]
data.columns = cleaned_column_names
data.head()
```

```python
if 'nan' in data.columns:
    data = data.drop(columns=['nan'])
data.head()
```

```python
data.columns.tolist()
```

```python
column_mapping = {
    'geo_id': 'geo_id',
    'block_group': 'block_group',
    'census_tract': 'census_tract',
    'counties': 'county',
    'state': 'state',
    'estimate_total': 'est_total',
    'margin_of_error_total': 'moe_total',
    'estimate_total_housing_units_with_a_mortgage_contract_to_purchase_or_similar_debt': 'est_mortgage_debt',
    'margin_of_error_total_housing_units_with_a_mortgage_contract_to_purchase_or_similar_debt': 'moe_mortgage_debt',
    'estimate_total_housing_units_with_a_mortgage_contract_to_purchase_or_similar_debt_housing_units_with_no_second_mortgage_and_no_home_equity_loan': 'est_no_second_mortgage_no_loan',
    'margin_of_error_total_housing_units_with_a_mortgage_contract_to_purchase_or_similar_debt_housing_units_with_no_second_mortgage_and_no_home_equity_loan': 'moe_no_second_mortgage_no_loan',
    'estimate_total_housing_units_with_a_mortgage_contract_to_purchase_or_similar_debt_housing_units_with_multiple_mortgages': 'est_multiple_mortgages',
    'margin_of_error_total_housing_units_with_a_mortgage_contract_to_purchase_or_similar_debt_housing_units_with_multiple_mortgages': 'moe_multiple_mortgages',
    'estimate_total_housing_units_with_a_mortgage_contract_to_purchase_or_similar_debt_housing_units_with_multiple_mortgages_mortgage_with_both_second_mortgage_and_home_equity_loan': 'est_both_second_mortgage_and_loan',
    'margin_of_error_total_housing_units_with_a_mortgage_contract_to_purchase_or_similar_debt_housing_units_with_multiple_mortgages_mortgage_with_both_second_mortgage_and_home_equity_loan': 'moe_both_second_mortgage_and_loan',
    'estimate_total_housing_units_with_a_mortgage_contract_to_purchase_or_similar_debt_housing_units_with_multiple_mortgages_mortgage_with_only_home_equity_loan': 'est_only_home_equity_loan',
    'margin_of_error_total_housing_units_with_a_mortgage_contract_to_purchase_or_similar_debt_housing_units_with_multiple_mortgages_mortgage_with_only_home_equity_loan': 'moe_only_home_equity_loan',
    'estimate_total_housing_units_with_a_mortgage_contract_to_purchase_or_similar_debt_housing_units_with_multiple_mortgages_mortgage_with_only_second_mortgage': 'est_only_second_mortgage',
    'margin_of_error_total_housing_units_with_a_mortgage_contract_to_purchase_or_similar_debt_housing_units_with_multiple_mortgages_mortgage_with_only_second_mortgage': 'moe_only_second_mortgage',
    'estimate_total_housing_units_with_a_mortgage_contract_to_purchase_or_similar_debt_home_equity_loan_without_a_primary_mortgage': 'est_home_equity_loan_only',
    'margin_of_error_total_housing_units_with_a_mortgage_contract_to_purchase_or_similar_debt_home_equity_loan_without_a_primary_mortgage': 'moe_home_equity_loan_only',
    'estimate_total_housing_units_without_a_mortgage': 'est_without_mortgage',
    'margin_of_error_total_housing_units_without_a_mortgage': 'moe_without_mortgage'
}

data.rename(columns=column_mapping, inplace=True)
print("Column names after renaming:")
print(data.columns)

numeric_columns = [
    'est_total',
    'moe_total',
    'est_mortgage_debt',
    'moe_mortgage_debt',
    'est_no_second_mortgage_no_loan',
    'moe_no_second_mortgage_no_loan',
    'est_multiple_mortgages',
    'moe_multiple_mortgages',
    'est_both_second_mortgage_and_loan',
    'moe_both_second_mortgage_and_loan',
    'est_only_home_equity_loan',
    'moe_only_home_equity_loan',
    'est_only_second_mortgage',
    'moe_only_second_mortgage',
    'est_home_equity_loan_only',
    'moe_home_equity_loan_only',
    'est_without_mortgage',
    'moe_without_mortgage'
]

missing_columns = [col for col in numeric_columns if col not in data.columns]
if missing_columns:
    print(f"These columns are missing after renaming: {missing_columns}")
else:
    data.replace(['**', '***'], np.nan, inplace=True)
    data[numeric_columns] = data[numeric_columns].replace({',': '', '-': np.nan, r'\+': ''}, regex=True)
    data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric, errors='raise')

data.fillna('', inplace=True)
if 'nan' in data.columns:
    data.drop(columns=['nan'], inplace=True)
data.head()
```
